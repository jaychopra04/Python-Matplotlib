```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([1,3,7,8,9])
y=np.array([21,27,31,43,45])

f1={'color':'green','size':'15'}
f2={'family':'serif','color':'hotpink','size':'10'}

plt.title("My perfect graph",fontdict=f1, loc='center')
plt.xlabel("No. of x coordinates",fontdict=f2)
plt.ylabel("No. of y coordinates",fontdict=f2)

plt.plot(x,y,marker='p',lw='2',c='orange')
plt.grid(color='green',lw='2',ls='dotted')
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
f1={'family':'serif','color':'hotpink','size':'12'}
# p1
x=np.array([1,2,3,4])
y=np.array([90,92,94,98])

plt.subplot(1,2,1)
plt.title('Academics')
plt.xlabel("----HI JAY")
plt.plot(x,y,marker='.')
plt.grid(ls=':',c='green')

# p2
x=np.array([1,2,3,4])
y=np.array([90,90,92,98])

plt.subplot(1,2,2)
plt.title('self')
plt.xlabel("Great----")
plt.plot(x,y,marker='D')
plt.grid(ls=':',c='r')

plt.suptitle("My progress graph",fontdict=f1)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
f1={'family':'serif','color':'hotpink','size':'12'}
# p1
x=np.array([1,2,3,4])
y=np.array([90,92,94,98])

plt.subplot(1,2,1)
plt.title('Academics')
plt.xlabel("----HI JAY")
plt.plot(x,y,marker='.')
plt.grid(ls=':',c='green')

# p2
x=np.array([1,2,3,4])
y=np.array([90,90,92,98])
clr=np.array(['green','blue','cyan','brown'])

plt.subplot(1,2,2)
plt.title('self')
plt.xlabel("Great----")
plt.scatter(x,y,c=clr)
plt.grid(ls=':',c='r')

plt.suptitle("My progress graph",fontdict=f1)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

a=np.array([1,2,3,4,5,6,6,7])
b=np.array([1,2,3,4,5,6,8,6])
d=np.array([2,34,56,78,23,89,37,99])

plt.scatter(a,b, c=d, cmap='viridis',alpha=1.4)

plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y=np.array([20,30,15,5,25,5])

des=['A','B','C','D','E','F']
plt.pie(y, labels=des)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y=np.array([20,30,15,5,25,5])

des=['A','B','C','D','E','F']
exp=[0.2,0,0,0.2,0,0]

plt.pie(y, labels=des, explode=exp, shadow=True)
plt.show()
```
